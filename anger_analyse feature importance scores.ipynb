```python
ROOT_PATH = '/content/drive/MyDrive/Colab Notebooks/Allen_NLP_hackathon'
feature_importance_scores_dir = os.path.join(ROOT_PATH, 'Results')
```

```python
def read_csv_files(path_to_dir):
    dataframes = []
    all_files = glob.glob(path_to_dir + "/*.csv")
    for filename in all_files:
        df = pd.read_csv(filename, index_col=None)
        dataframes.append(df)

    return dataframes
```

```python
data = pd.concat(read_csv_files(feature_importance_scores_dir))
```

```python
female_seed_words = ['sister', 'female', 'woman', 'girl', 'daughter', 'she', 'hers', 'her','mother']
male_seed_words = ['brother', 'male', 'man', 'boy', 'son', 'he', 'his', 'him','father']
christian_seed_words = ['christianity', 'christian', 'church', 'bible', 'priest']
muslim_seed_words = ['islam', 'muslim', 'mosque', 'quran', 'imam']
jew_seed_words = ['judaism', 'jew', 'synagogue', 'torah', 'rabbi']
```

```python
def find_average_attributions_for_seed_words(data, wordlist):
  words = []
  avergate_attributions = []
  seed_word_dic = {}
  for i in wordlist:
    words.append(i)
    word_attributions = data[data["words"] == i]
    avergate_attributions.append(np.mean(word_attributions.attribution))
  seed_word_dic["word"] = words
  seed_word_dic["avg_attribution_scores"] = avergate_attributions
  return pd.DataFrame(seed_word_dic)
```

```python
female_data_frmae = find_average_attributions_for_seed_words(data, female_seed_words)
```
